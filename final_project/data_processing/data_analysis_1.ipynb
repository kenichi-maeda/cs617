```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
combined_df = pd.read_csv('../data/combined.csv')
year_groups = combined_df.groupby('YEAR').size()

plt.figure(figsize=(10, 6))
year_groups.plot(kind='bar', color='skyblue')
plt.title('Number of Crime Incidents 2015-2024')
plt.xlabel('Year')
plt.ylabel('Number of Items')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# List of specific offense descriptions I want to filter out
incidents_to_remove = ['INVESTIGATE PERSON', 'SEARCH WARRANT', 'DRUGS', 'VAL', 'SICK', 'TOWED MOTOR VEHICLE', 
                    'WARRANT ASSIST', 'M/V', 'VIOL. OF RESTRAINING ORDER W NO ARREST', 'SERVICE TO OTHER PD INSIDE OF MA.', 
                    'FIRE', 'DEATH INVESTIGATION', 'MISSING PERSON', 'FORGERY / COUNTERFEITING', 'ANIMAL',
                    'RECOVERED', 'VIOL. OF RESTRAINING ORDER W ARREST', 'LANDLORD', 'INVESTIGATION FOR ANOTHER AGENCY'
                    'CONSPIRACY EXCEPT DRUG LAW', 'OPERATING UNDER THE INFLUENCE ALCOHOL', 'REPORT AFFECTING OTHER DEPTS.',
                    'SERVICE TO OTHER PD OUTSIDE OF MA.', 'AIRCRAFT INCIDENTS', 'GAMBLING - BETTING / WAGERING', 
                    'COUNTERFEITING', 'PROTECTIVE CUSTODY / SAFEKEEPING', 'Migrated Report', 'SERVICE TO OTHER AGENCY',
                    'OPERATING UNDER THE INFLUENCE (OUI) ALCOHOL', 'DRUNKENNESS', 'CHILD REQUIRING ASSISTANCE (FOMERLY CHINS)'
                    'Evidence Tracker Incidents', 'WARRANT ARREST', 'EVADING FARE', 'LIQUOR', 'SUDDEN DEATH'] 

# Create a new DataFrame containing only rows with specific offense descriptions
only_crimes_df = combined_df2[~combined_df2['OFFENSE_DESCRIPTION'].isin(incidents_to_remove)]
top_values = only_crimes_df['OFFENSE_DESCRIPTION'].value_counts().head(50)

plt.figure(figsize=(12, 12))
top_values.plot(kind='bar', color='skyblue')
plt.title('After filtering')
plt.xlabel('Incident Description')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
year_groups = only_crimes_df.groupby('YEAR').size()

plt.figure(figsize=(10, 6))
year_groups.plot(kind='bar', color='skyblue')
plt.title('Number of Crime Incidents 2015-2024')
plt.xlabel('Year')
plt.ylabel('Number of Items')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
