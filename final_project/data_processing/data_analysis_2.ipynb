```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/only_crimes_.csv')
```

```python
df['OCCURRED_ON_DATE'] = pd.to_datetime(df['OCCURRED_ON_DATE'])

year_groups = df.groupby('YEAR').size().rename('count')

print(year_groups)

years_to_exclude = [2015, 2024]
filtered_year_groups = year_groups.drop(years_to_exclude, errors='ignore')


filtered_year_groups.to_csv('../data/numberOfCrimesOverYearsTotal.csv', header=True)

plt.figure(figsize=(10, 6))
filtered_year_groups.plot(kind='bar', color='skyblue')
plt.title('Number of Crime Incidents 2015-2024')
plt.xlabel('Year')
plt.ylabel('Number of Items')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
assault_df = df[df['OFFENSE_DESCRIPTION'] == "ASSAULT"]
assault_count_by_year = assault_df.groupby('YEAR').size()

years_to_exclude = [2015, 2024]
filtered_assault_groups = assault_count_by_year.drop(years_to_exclude, errors='ignore')


filtered_assault_groups.to_csv('../data/numberOfCrimesOverYearsAssault.csv', header=True)

plt.figure(figsize=(10, 6))
filtered_assault_groups.plot(kind='bar')


plt.xticks(range(len(filtered_assault_groups.index)), filtered_assault_groups.index)
plt.title('Total Assault Incidents Per Year')
plt.xlabel('Year')
plt.ylabel('Total Incidents')
plt.grid(True)
plt.show()
```
