```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/only_crimes_.csv')
df['OCCURRED_ON_DATE'] = pd.to_datetime(df['OCCURRED_ON_DATE'])
```

```python
df['YEAR'] = df['OCCURRED_ON_DATE'].dt.year

# Important!!!!!!!!!
# Exclude 00:00:00 since there is an unusual peak (might be due to administrative tasks)
df_2023 = df[(df['YEAR'] == 2023) & (df['OCCURRED_ON_DATE'].dt.strftime('%H:%M:%S') != '00:00:00')].copy()
df_2023['time'] = df_2023['OCCURRED_ON_DATE'].dt.strftime('%H:%M')

# Group by 'time'
grouped_by_time = df_2023.groupby('time').size()

time_index = pd.date_range("00:00", "23:59", freq="T").strftime('%H:%M')
grouped_by_time = grouped_by_time.reindex(time_index, fill_value=0)


df_to_save = pd.DataFrame({'time': grouped_by_time.index, 'count': grouped_by_time.values})
df_to_save.to_csv('../data/daytime_total_data.csv', index=False)


plt.figure(figsize=(18, 6))
plt.plot(grouped_by_time.index, grouped_by_time.values, label='Incidents by Time', color='dodgerblue')

hour_ticks = [f"{hour:02}:00" for hour in range(24)]
plt.xticks(hour_ticks, hour_ticks, rotation=90)

plt.title('Number of Incidents by Specific Time in 2023')
plt.xlabel('Time of Day (Hour)')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.tight_layout()
plt.show()
```
